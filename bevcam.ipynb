```python
import importlib
import os
import sys
import tqdm
import numpy as np
import cv2
import tensorflow as tf
# import configargparse
import matplotlib.pyplot as plt
import utils


# determine absolute filepaths
one_hot_palette_input  = utils.abspath('/content/drive/MyDrive/bev/cam2bev/model/one_hot_conversion/my_convert_10.xml')
one_hot_palette_label  = utils.abspath('/content/drive/MyDrive/bev/cam2bev/model/one_hot_conversion/my_convert_3.xml')
model_bev              = utils.abspath('/content/drive/MyDrive/bev/cam2bev/model/architecture/uNetXST.py')
unetxst_homographies   = utils.abspath('/content/drive/MyDrive/bev/cam2bev/preprocessing/homography_converter/uNetXST_homographies/2_F.py')
model_weights          = utils.abspath('/content/drive/MyDrive/bev/cam2bev/model/output/2023-04-25-23-12-05/Checkpoints/e003_weights.hdf5')


# load network architecture module
architecture = utils.load_module(model_bev)


# get max_samples_testing samples
n_inputs = 1  # only use front image
n_samples = 1  # number of front images
image_shape_original = (1216, 1936)


# parse one-hot-conversion.xml
one_hot_palette_input = utils.parse_convert_xml(one_hot_palette_input)
one_hot_palette_label = utils.parse_convert_xml(one_hot_palette_label)
n_classes_input = len(one_hot_palette_input)
n_classes_label = len(one_hot_palette_label)


# build model
image_shape = [256, 512]
if unetxst_homographies is not None:
  uNetXSTHomographies = utils.load_module(unetxst_homographies)
  model_bev = architecture.get_network((image_shape[0], image_shape[1], n_classes_input), n_classes_label, n_inputs=n_inputs, thetas=uNetXSTHomographies.H)
else:
  model_bev = architecture.get_network((image_shape[0], image_shape[1], n_classes_input), n_classes_label)
model_bev.load_weights(model_weights)
print(f"Reloaded model from {model_weights}")
```

```python
def inference_bev(model, img):
  # input_sequence: (n_samples, image shape)
  input_sequence = []
  test_img = np.array(img)
  input_sequence.append(test_img)
  input_sequence = np.asarray(input_sequence)

  # build data parsing function
  def parse_sample(inp):
      # parse and process input images
      input = []
      # inp is an image of size image_shape_original in cv2.COLOR_BGR2RGB
      inp = np.squeeze(inp)
      inp = utils.resize_image_op(inp, image_shape_original, image_shape, interpolation=tf.image.ResizeMethod.NEAREST_NEIGHBOR)
      inp = utils.one_hot_encode_image_op(inp, one_hot_palette_input)
      input.append(inp)
      input = input[0]
      return input

  # run predictions
  # print(f"Running predictions...")
  for k in tqdm.tqdm(range(n_samples)):

      # load sample
      input = parse_sample(input_sequence[k, ])

      # add batch dim
      input = np.expand_dims(input, axis=0)  # n_inputs == 1

      # run prediction
      prediction = model.predict(input).squeeze()

      # convert to output image
      prediction = utils.one_hot_decode_image(prediction, one_hot_palette_label)
      return prediction
```

```python
def getLines(img):

    # Define the color ranges for each color of interest for creating masks.
    COLOR1_RANGE = [(30, 0, 0), (255, 50, 50)]  # Blue in BGR, [(low), (high)].
    COLOR2_RANGE = [(0, 30, 0), (50, 255, 50)] 

    # Create masks:
    color1_mask = cv2.inRange(img, COLOR1_RANGE[0], COLOR1_RANGE[1])
    color2_mask = cv2.inRange(img, COLOR2_RANGE[0], COLOR2_RANGE[1])

    # Adjust according to your adjacency requirement.
    kernel = np.ones((3, 3), dtype=np.uint8)

    # Dilating masks to expand boundary.
    color1_mask = cv2.dilate(color1_mask, kernel, iterations=1)
    color2_mask = cv2.dilate(color2_mask, kernel, iterations=1)

    # Required points now will have both color's mask val as 255.
    common = cv2.bitwise_and(color1_mask, color2_mask)
    return common
```
